## Visualize Results

```python
labels = [ 'Logistic Regression', 'KNN', 'Naive Bayes', 'SVM', 'Decision Tree', 'Random Forest', 'XGBoost', 'Neural Network']
ds1 = [accs_lr[0], accs_knn[0], accs_nb[0], accs_svm[0], accs_dt[0], accs_rf[0], accs_xgb[0], accs_nn[0]]
ds2 = [accs_lr[1], accs_knn[1], accs_nb[1], accs_svm[1], accs_dt[1], accs_rf[1], accs_xgb[1], accs_nn[1]]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 5))
rects1 = ax.barh(x - width/2, ds1, width, label='No resampled')
rects2 = ax.barh(x + width/2, ds2, width, label='Resampled')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Accuracy (%)')
ax.set_title('So sánh Accuracy của các thuật toán sử dụng thư viện')
ax.set_yticks(x, labels)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

ax.bar_label(rects1, padding=3, fmt='%.2f')
ax.bar_label(rects2, padding=3, fmt='%.2f')

fig.tight_layout()

plt.show()
```

```python
labels = [ 'Logistic Regression', 'KNN', 'Naive Bayes', 'SVM', 'Decision Tree', 'Random Forest', 'XGBoost', 'Neural Network']
ds1 = [recalls_lr[0][0], recalls_knn[0][0]*100, recalls_nb[0][0]*100, recalls_svm[0][0]*100, recalls_dt[0][0]*100, recalls_rf[0][0]*100, recalls_xgb[0][0]*100, recalls_nn[0][0]*100]
ds2 = [recalls_lr[0][1], recalls_knn[0][1]*100, recalls_nb[0][1]*100, recalls_svm[0][1]*100, recalls_dt[0][1]*100, recalls_rf[0][1]*100, recalls_xgb[0][1]*100, recalls_nn[0][1]*100]
ds3 = [recalls_lr[1][0], recalls_knn[1][0]*100, recalls_nb[1][0]*100, recalls_svm[1][0]*100, recalls_dt[1][0]*100, recalls_rf[1][0]*100, recalls_xgb[1][0]*100, recalls_nn[1][0]*100]
ds4 = [recalls_lr[1][1], recalls_knn[1][1]*100, recalls_nb[1][1]*100, recalls_svm[1][1]*100, recalls_dt[1][1]*100, recalls_rf[1][1]*100, recalls_xgb[1][1]*100, recalls_nn[1][1]*100]

x = np.arange(len(labels))  # the label locations

fig, ax = plt.subplots(figsize=(10, 6))

# Plot recall for label 0
ax.plot(labels, ds1, marker='x', linestyle='-', label='Recall 0 (No resampled)')
ax.plot(labels, ds2, marker='s', linestyle='--', label='Recall 1 (No resampled)')

# Plot recall for label 1
ax.plot(labels, ds3, marker='o', linestyle='-', label='Recall 0 (Resampled)')
ax.plot(labels, ds4, marker='d', linestyle='--', label='Recall 1 (Resampled)')

# Add titles and labels
ax.set_ylabel('Recall (%)', fontsize=12)
ax.set_title('So sánh Recall cho nhãn 0 và nhãn 1 trên tập No resampled và Resampled', fontsize=14)
ax.legend(loc='best', fontsize=10)
ax.grid(True)

plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```
